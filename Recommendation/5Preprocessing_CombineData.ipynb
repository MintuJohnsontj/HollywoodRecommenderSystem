```python
import pandas as pd
```

# Read all Data

```python
# Movies untill 2016
df1 = pd.read_csv("prep1data.csv")
df1
```

```python
# Movies of 2017
df2 = pd.read_csv("prep2data.csv")
df2
```

```python
# Movies of 2018 to 2020
df3 = pd.read_csv("prep3data.csv")
df3
```

```python
# Movies of 2021 (Until June)
df4 = pd.read_csv("prep4data.csv")
df4
```

# Combining all Data

```python
df = df1.append(df2.append(df3.append(df4)))
df
```

```python
df['comb'] = df['actor_1_name'] + ' ' + df['actor_2_name'] + ' '+ df['actor_3_name'] + ' '+ df['director_name'] +' ' + df['genres']
```

```python
df
```

```python
df.drop_duplicates(subset ="movie_title", keep = 'last', inplace = True)
df
```

# Save the File

```python
df.to_csv('main_data.csv',index=False)
```
